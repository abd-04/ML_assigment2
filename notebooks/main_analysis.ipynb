Glucose-related biomarkers such as fasting glucose, postprandial glucose, and HbA1c show strong positive correlation with diabetes diagnosis. These variables are expected to be strong predictors for classification models.

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report, accuracy_score, mean_squared_error

features = [
    "glucose_fasting",
    "glucose_postprandial",
    "hba1c",
    "insulin_level",
    "bmi"
]

X_binary = df[features]
y_binary = df["diagnosed_diabetes"]

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X_binary,
    y_binary,
    test_size=0.2,
    random_state=42,
    stratify=y_binary
)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier()
}

for name, model in models.items():

    model.fit(X_train_b, y_train_b)

    preds = model.predict(X_test_b)

    print("\n", name)
    print(classification_report(y_test_b, preds))

    X_multi = df[features]
y_multi = df["diabetes_stage"]

X_train_m, X_test_m, y_train_m, y_test_m = train_test_split(
    X_multi,
    y_multi,
    test_size=0.2,
    random_state=42,
    stratify=y_multi
)

models_multi = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier()
}

for name, model in models_multi.items():

    model.fit(X_train_m, y_train_m)

    preds = model.predict(X_test_m)

    print("\n", name)
    print("Accuracy:", accuracy_score(y_test_m, preds))

    X_reg = df[features]
y_reg = df["diabetes_risk_score"]

X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reg,
    y_reg,
    test_size=0.2,
    random_state=42
)

reg_model = LinearRegression()

reg_model.fit(X_train_r, y_train_r)

preds = reg_model.predict(X_test_r)

rmse = mean_squared_error(y_test_r, preds, squared=False)

print("RMSE:", rmse)
```

```python
best_binary_model = LogisticRegression(max_iter=1000)
import joblib
best_binary_model.fit(X_train_b, y_train_b)
joblib.dump(best_binary_model, "../models/binary_model.pkl")


multi_model = LogisticRegression(max_iter=1000)

multi_model.fit(X_train_m, y_train_m)

joblib.dump(multi_model, "../models/multiclass_model.pkl")

from sklearn.linear_model import LinearRegression

reg_model = LinearRegression()

reg_model.fit(X_train_r, y_train_r)

joblib.dump(reg_model, "../models/regression_model.pkl")
```
